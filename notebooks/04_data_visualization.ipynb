```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Reproducibility
np.random.seed(42)

# Pretty defaults
plt.rcParams.update({
    'figure.dpi': 100,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'font.size': 11,
})
```

## 1. Line Plot — Trends Over Time

```python
months = np.arange(1, 13)
revenue = [42, 38, 55, 61, 70, 66, 74, 80, 77, 85, 91, 105]
cost    = [30, 32, 35, 38, 40, 41, 43, 45, 44, 47, 50, 55]

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(months, revenue, marker='o', label='Revenue ($k)', color='steelblue', linewidth=2)
ax.plot(months, cost, marker='s', linestyle='--', label='Cost ($k)', color='tomato', linewidth=2)
ax.fill_between(months, cost, revenue, alpha=0.1, color='green', label='Profit margin')
ax.set_title('Monthly Revenue vs Cost')
ax.set_xlabel('Month')
ax.set_ylabel('Amount ($k)')
ax.set_xticks(months)
ax.legend()
plt.tight_layout()
plt.show()
```

## 2. Scatter Plot — Relationships Between Variables

```python
n = 120
height = np.random.normal(170, 10, n)
weight = 0.55 * height - 20 + np.random.normal(0, 5, n)
gender = np.random.choice(['Male','Female'], n)

colors = ['#3498DB' if g=='Male' else '#E91E63' for g in gender]

fig, ax = plt.subplots(figsize=(7, 5))
scatter = ax.scatter(height, weight, c=colors, alpha=0.7, edgecolors='white', s=60)
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Weight (kg)')
ax.set_title('Height vs Weight')

from matplotlib.patches import Patch
legend_elements = [Patch(color='#3498DB', label='Male'),
                   Patch(color='#E91E63', label='Female')]
ax.legend(handles=legend_elements)

# Add regression line
coeffs = np.polyfit(height, weight, 1)
x_line = np.linspace(height.min(), height.max(), 100)
ax.plot(x_line, np.polyval(coeffs, x_line), 'k--', alpha=0.5, label='Trend')
plt.tight_layout()
plt.show()
```

## 3. Bar and Horizontal Bar Charts — Comparisons

```python
categories = ['Python','JavaScript','Java','C++','R','Go']
counts = [55, 47, 35, 28, 18, 12]

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Vertical bar
bar_colors = plt.cm.Blues(np.linspace(0.4, 0.9, len(categories)))
axes[0].bar(categories, counts, color=bar_colors, edgecolor='white')
axes[0].set_title('Popularity by Language (vertical)')
axes[0].set_ylabel('Survey respondents (%)')
for i, v in enumerate(counts):
    axes[0].text(i, v+0.5, str(v), ha='center', fontweight='bold')

# Horizontal bar — better for long labels
sorting = np.argsort(counts)
axes[1].barh([categories[i] for i in sorting], [counts[i] for i in sorting],
             color=bar_colors)
axes[1].set_title('Popularity by Language (horizontal)')
axes[1].set_xlabel('Survey respondents (%)')
plt.tight_layout()
plt.show()
```

## 4. Histogram and Density Plot — Distributions

```python
scores = np.concatenate([
    np.random.normal(65, 8, 80),
    np.random.normal(85, 5, 40)
])

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

# Histogram
axes[0].hist(scores, bins=20, color='steelblue', edgecolor='white', alpha=0.8)
axes[0].axvline(np.mean(scores), color='red', linestyle='--', label=f'Mean={np.mean(scores):.1f}')
axes[0].axvline(np.median(scores), color='orange', linestyle='--', label=f'Median={np.median(scores):.1f}')
axes[0].set_xlabel('Score')
axes[0].set_ylabel('Count')
axes[0].set_title('Score Distribution')
axes[0].legend()

# Box plot
group_a = np.random.normal(70, 10, 50)
group_b = np.random.normal(80, 6, 50)
group_c = np.random.normal(75, 12, 50)
axes[1].boxplot([group_a, group_b, group_c],
                labels=['Group A','Group B','Group C'],
                patch_artist=True,
                boxprops=dict(facecolor='steelblue', alpha=0.5))
axes[1].set_title('Score Distribution by Group')
axes[1].set_ylabel('Score')
plt.tight_layout()
plt.show()
```

## 5. Heatmap — Correlation Matrix

```python
# Generate a correlation-like matrix
np.random.seed(0)
raw = np.random.randn(100, 5)
cols = ['Age','Income','Score','Experience','Satisfaction']
df = pd.DataFrame(raw, columns=cols)
corr = df.corr()

fig, ax = plt.subplots(figsize=(6, 5))
im = ax.imshow(corr, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(im, ax=ax)
ax.set_xticks(range(len(cols))); ax.set_xticklabels(cols, rotation=45, ha='right')
ax.set_yticks(range(len(cols))); ax.set_yticklabels(cols)
for i in range(len(cols)):
    for j in range(len(cols)):
        ax.text(j, i, f'{corr.iloc[i,j]:.2f}', ha='center', va='center', fontsize=9)
ax.set_title('Correlation Matrix')
plt.tight_layout()
plt.show()
```

## 6. The Anatomy of a Good Plot

A well-designed plot has:
1. **Title** — what the chart shows
2. **Axis labels with units** — never leave an axis unlabelled
3. **Legend** — if multiple series
4. **Appropriate scale** — start y-axis at 0 for bar charts
5. **Minimal chartjunk** — remove gridlines and borders that do not add information
6. **Accessible colours** — colour-blind friendly palettes (e.g., `viridis`, `tab10`)

```python
# Demonstrating good vs bad practices
fig, axes = plt.subplots(1, 2, figsize=(13, 4))
x = ['Q1','Q2','Q3','Q4']
y = [100, 102, 104, 103]

# 'Bad' chart — truncated y axis exaggerates small differences
axes[0].bar(x, y, color='gray')
axes[0].set_ylim(98, 106)
axes[0].set_title('Misleading: y-axis starts at 98')
axes[0].set_ylabel('Sales')

# 'Good' chart — zero baseline
axes[1].bar(x, y, color='steelblue')
axes[1].set_ylim(0, 120)
axes[1].set_title('Honest: y-axis starts at 0')
axes[1].set_ylabel('Sales')

plt.tight_layout()
plt.show()
print('Same data — very different impression!')
```

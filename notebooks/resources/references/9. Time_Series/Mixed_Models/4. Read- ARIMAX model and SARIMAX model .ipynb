# ARIMAX model and SARIMAX model


## Prerequisites:

To start ARIMAX and SARIMAX models, you must have proper understanding of a few basic concepts like:

- Time Series data  
- Auto-regressive (AR) model
- Moving Average (MA) model
- Auto-regressive Moving Average (ARMA) model
- Auto-regressive Integrated Moving Average (ARIMA) model
- Seasonal Auto-regressive Integrated Moving Average (SARIMA) model


## Learning Objective:

- Understand the concept of endogenous and exogenous variables.

- Understand the concepts of ARIMAX and SARIMAX models.


## Endogenous and Exogenous Variables

In time series analysis, we have studied about models that forecast the values for a variable $(X)$ based on its past values (auto-regressive) and random noises $(Z)$ so far. We have not introduced any independent variable (say $U$) in our models such that the value of $(X)$ at time $t$ (i.e. $X_t$)depends on $U_t$. This independent variable $(U_t)$ must not be random noise like $Z_t$. However, some other process or model (for example, another linear regression model) should generate $U_t$. Note that our TSA model won't determine $U_t$, so it's exogenous to our model. Since the variable $(X)$ is the one the model determines, $X$ is the endogenous variable.


## ARIMAX model

ARIMAX stands for Auto-Regressive Integrated Moving Average with eXogenous variable. As the name suggests, an additional exogenous variable is introduced in the ARIMA model for forecasting the endogenous variable.

\\

### Mathematical Expression for $\text{ARIMAX} (p, d, q)$ Model

>$ \phi_p(B) \nabla ^d X_t = \tau U_t + \theta_q(B) Z_t $


where,
* $X_t$ is endogenous variable for the ARIMAX model at time $t$.

* $U_t$ is exogenous variable for the model at time $t$.

* $Z_t$ is white noise introduced at time $t$.


* $p$, $d$ and $q$ are the orders for AR-part, differencing and MA-part.

* $\phi_p(B) = 1 - \sum_{i=1} ^p \phi_i B^i$ is operation for AR-part. 

* $\theta_q(B) = 1 + \sum_{i = 1} ^q \theta_i B^i$ is operation for MA-part.

* $\tau$ is the parameter for the exogenous variable.

* $\nabla$ is a difference operator, $\nabla = 1 - B$ where $B$ is a backward operator. 
    * $\nabla X_t = (1- B) X_t = X_t - X_{t-1}$



## SARIMAX model
SARIMAX stands for Seasonal Auto-Regressive Integrated Moving Average with eXogenous variable. The SARIMAX model extends the SARIMA model with the introduction of a separate exogenous variable (say $U_t$). 

\\

### Mathematical Expression for $\text{ARIMAX} (p, d, q) \times (P, D, Q)_s$ Model

>$
\Phi_P(B^s) \phi_p(B) \nabla_s ^D \nabla ^d X_t = \tau U_t + \Theta_Q(B^s) \theta_q(B) Z_t
$


where,

* $X_t$ is endogenous variable for the ARIMAX model at time $t$.

* $U_t$ is exogenous variable for the model at time $t$.

* $Z_t$ is white noise introduced at time $t$.


* $p$, $d$ and $q$ are the orders for non-seasonal AR-part, differencing and MA-part.

* $\phi_p(B)$ and $\theta_q(B)$ are non-seasonal AR and MA operators respectively.

* $P$, $D$, and $Q$ are the orders for seasonal AR-part, differencing, and MA-part.

* $\tau$ is the parameter for the exogenous variable.


* $s$ is the season for the model.

* $\Phi_P(B^s)$ and $\Theta_Q(B^s)$ are seasonal AR and MA operators.


* $\nabla _s = 1 - B^s$ is the seasonal difference operator.
    * $\nabla _s X_t = (1- B^s) X_t = X_t - X_{t - s}$ 


## Key Takeaways

1. The endogenous variable is determined by the Time Series model, but the exogenous variable is not determined by the model but by some other exterior process or model. 

2. The ARIMAX model extends the ARIMA model by introducing an exogenous variable, whereas the SARIMAX model extends the SARIMA model.
